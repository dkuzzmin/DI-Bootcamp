```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# 1. Data Preparation:

```python
array = np.random.uniform(low=-5, high=35, size=(10,12))

cities = ['Reykyavik', 'Oslo', 'Tromse', 'Amsterdam', 'Malaga', 'Barcelona', 'Nice', 'Cannes', 'Turin', 'Milan']
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

df = pd.DataFrame(array, index=cities, columns=months)
```

```python
df
```

# 2. Data Analysis:

```python
avg_temp = df.mean(axis=1)
print(avg_temp)
```

```python
city_max_temp = avg_temp.argmax()
df.index[city_max_temp]
```

```python

city_min_temp = avg_temp.argmin()
df.index[city_min_temp]
```

# 3. Data Visualization:

```python
plt.figure(figsize=(20, 8))

for city in df.index:
    if city == df.index[city_max_temp]:
        plt.plot(df.columns, df.loc[city], label=city, linewidth=4, linestyle=':', color='green')
    elif city == df.index[city_min_temp]:
        plt.plot(df.columns, df.loc[city], label=city, linewidth=4, linestyle='--', color='blue')
    else:
        plt.plot(df.columns, df.loc[city], label=city, linestyle='-', alpha=0.7)

plt.title("Monthly Temperature Trends for Each City")
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.legend()
plt.grid(True)
plt.show()
```

According to my data, the coldest city on average is Malaga, and the warmest city on average per year is Nice
