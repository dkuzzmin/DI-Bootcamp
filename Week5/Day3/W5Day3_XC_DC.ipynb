```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Use the ‘Iris’ dataset available in seaborn

```python
df = sns.load_dataset("iris")
df
```

# Create a scatter plot of sepal length vs. sepal width, color-coded by species.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

# Generate a histogram for petal lengths.

```python



plt.hist(df['petal_length'], bins=20, color='blue')
plt.title('Petal Lengths')
plt.xlabel('Petal Length, sm')
plt.ylabel('Frequency')
plt.show()
```

# Produce a pair plot to show pairwise relationships in the dataset


```python
sns.pairplot(df, hue='species')




plt.suptitle("Relationships 'One-to-One'", y=1.1)
plt.show()
```

# Create an interactive 3D scatter plot mapping sepal length, sepal width, and petal length, color-coded by species.

```python
import plotly.express as px
```

```python
fig = px.scatter_3d(df, x='sepal_length', y='sepal_width', z='petal_length', color='species',
                    title='3D Scatter Plot')

fig.show()
```
