**Exercise 2 : Bar Plot Of Categorical Data With Plotnine**

```python
df = pd.DataFrame({'sport': np.random.choice(['footbal', 'basketball', 'hockey'], 30)})
df.head()
```

```python
s_count = df['sport'].value_counts()
s_count
```

```python
plt.bar(s_count.index, s_count.values)
plt.title('Category Frequency')
plt.xlabel('Sport')
plt.ylabel('Amount')


plt.show()
```

**Exercise 3 : Line Plot With Multiple Lines With Plotnine**

```python
x = np.linspace(0, 10, 500)
y1 = np.sin(x)
y2 = np.cos(x)
df = pd.DataFrame({'X': x, 'sin(X)': y1, 'cos(X)': y2})
```

```python
sns.lineplot(x='X', y='sin(X)', data=df, label='sin(X)', color='green')
sns.lineplot(x='X', y='cos(X)', data=df, label='cos(X)', color='black')

plt.xlabel('X-Value')
plt.ylabel('Y-Value')

plt.title('Comparison of Two Functions')

plt.show()
```
