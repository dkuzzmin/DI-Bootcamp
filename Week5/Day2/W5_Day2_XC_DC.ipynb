**2. Data Visualization with Plotly:**
- Create an interactive line chart to show sales trends over the years.
- Build an interactive map to visualize sales distribution by states.

```python
df['Year'] = pd.DatetimeIndex(df['Order Date']).year
df.head()
```

```python
ndf = df.groupby('Year')['Sales'].sum()
ndf.name = 'Sales'
ndf = ndf.reset_index()
ndf
```

```python

# sns.lineplot(data=ndf, x='Year', y='Sales')
# sns.set_style('darkgrid')
# plt.title('Sales trend in 2014-2017')

# plt.xticks(ndf['Year'])
# plt.show()



fig = px.line(ndf, x='Year', y='Sales', title = 'Sales trend in 2014-2017')


fig.update_xaxes(dtick=1)

fig.show()
```

```python
sdf = df.groupby('State')['Sales'].sum()
sdf.name = 'Sales'
sdf = sdf.reset_index()
sdf.head()
```

```python
shortstate = {
    'Alabama': 'AL',
    'Alaska': 'AK',
    'Arizona': 'AZ',
    'Arkansas': 'AR',
    'California': 'CA',
    'Colorado': 'CO',
    'Connecticut': 'CT',
    'Delaware': 'DE',
    'Florida': 'FL',
    'Georgia': 'GA',
    'Hawaii': 'HI',
    'Idaho': 'ID',
    'Illinois': 'IL',
    'Indiana': 'IN',
    'Iowa': 'IA',
    'Kansas': 'KS',
    'Kentucky': 'KY',
    'Louisiana': 'LA',
    'Maine': 'ME',
    'Maryland': 'MD',
    'Massachusetts': 'MA',
    'Michigan': 'MI',
    'Minnesota': 'MN',
    'Mississippi': 'MS',
    'Missouri': 'MO',
    'Montana': 'MT',
    'Nebraska': 'NE',
    'Nevada': 'NV',
    'New Hampshire': 'NH',
    'New Jersey': 'NJ',
    'New Mexico': 'NM',
    'New York': 'NY',
    'North Carolina': 'NC',
    'North Dakota': 'ND',
    'Ohio': 'OH',
    'Oklahoma': 'OK',
    'Oregon': 'OR',
    'Pennsylvania': 'PA',
    'Rhode Island': 'RI',
    'South Carolina': 'SC',
    'South Dakota': 'SD',
    'Tennessee': 'TN',
    'Texas': 'TX',
    'Utah': 'UT',
    'Vermont': 'VT',
    'Virginia': 'VA',
    'Washington': 'WA',
    'West Virginia': 'WV',
    'Wisconsin': 'WI',
    'Wyoming': 'WY'
}


sdf['State'] = sdf['State'].map(shortstate)
sdf
```

```python

fig = px.choropleth(
    sdf,
    locations='State',
    locationmode="USA-states",
    color='Sales',
    hover_name='State',
    scope="usa",
    color_continuous_scale=px.colors.sequential.Plasma
    )
fig.update_layout(title_text='Sales Distribution by states')
fig.show()
```

**3. Data Visualization with Plotnine:**

```python
vdf = df[['Product Name', 'Product ID', 'Sales']]
vdf.head()
```

```python
vdf10 = vdf.groupby('Product ID')['Sales'].sum()
vdf10.name = 'Sales'
vdf10 = vdf10.reset_index()
vdf10
vdf10 = vdf10.nlargest(10, 'Sales')
vdf10
```

```python
plt.figure(figsize=(10, 6))
plt.bar(vdf10['Product ID'], vdf10['Sales'], color = 'red')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Top 10 Products')
plt.xticks(rotation=90)
plt.show()
```
