**6.Social Support and Happiness:
Employ a Matplotlib scatter plot to investigate the relationship between ‘Social support’ and ‘Score’.**

```python
x = df['Social support']
y = df['Score']


plt.scatter(x,y)
plt.title('Scatter Plot with Social Support and Happiness')
plt.xlabel('Social Support')
plt.ylabel('Score')

plt.show()
```

**7.Regional Comparisons of GDP and Health:**

**8.Create a Matplotlib subplot that compares ‘GDP per Capita’ and ‘Healthy Life Expectancy’ across different regions.**

```python
y1 = [df[df["Country or region"] == 'Israel']['GDP per capita'].values[0],
df[df["Country or region"] == 'Russia']['GDP per capita'].values[0],
df[df["Country or region"] == 'United States']['GDP per capita'].values[0]]
y2 = [df[df["Country or region"] == 'Israel']['Healthy life expectancy'].values[0],
df[df["Country or region"] == 'Russia']['Healthy life expectancy'].values[0],
df[df["Country or region"] == 'United States']['Healthy life expectancy'].values[0]]
x = ['Israel', 'Russia', 'United States']
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))


# First subplot
ax1.bar(x, y1)
ax1.set_title('GDP per capita')

# Second subplot
ax2.bar(x, y2)
ax2.set_title('Healthy life expectancy')

plt.subplots_adjust(wspace=0.4)

plt.show()
```

**9.Use a bar plot for ‘GDP per Capita’ and a line plot for ‘Healthy Life Expectancy’ on the same axes to observe how economic strength relates to health outcomes in different regions.**

```python
plt.figure(figsize=(18, 5))
x = df['Country or region']
y1 = df['GDP per capita']
y2 = df['Healthy life expectancy']

plt.bar(x, y1, color='green', label='GDP per Capita')
plt.plot(x, y2, color='red', label='Healthy Life Expectancy')
plt.title('GDP per Capita and Healthy Life Expectancy')
plt.xlabel('Country or region')
plt.ylabel('Value')

plt.legend()
plt.show()
```
