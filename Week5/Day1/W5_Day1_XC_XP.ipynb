**Exercise 3 : Visualizing Monthly Sales With A Bar Chart**

```python
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = np.random.normal(12000, 3000, 12)
sales
```

```python
plt.bar(month, sales, hatch='*')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales Amount($)')
plt.show()
```

**Exercise 4 : Data Visualisation**

```python
df = pd.read_csv('sales_data.csv')
df
```

Calculate the total quantity of products sold by the company.


```python
df['quantity'].sum()
```

Identify the category that had the highest revenue and calculate how much revenue it generated.


```python
df.groupby('category')['revenue'].sum().max()
```

Determine the average revenue per sale in the dataset.


```python
round(df['revenue'].mean())
```

Create a bar chart to visualize the total revenue generated in each quarter of the year (Q1, Q2, Q3, Q4).


```python
q1 = df[('2022-01-01' <= df['date']) & (df['date'] <= '2022-03-31')]['revenue'].sum()
q2 = df[('2022-04-01' <= df['date']) & (df['date'] <= '2022-06-30')]['revenue'].sum()
q3 = df[('2022-07-01' <= df['date']) & (df['date'] <= '2022-09-30')]['revenue'].sum()
q4 = df[('2022-10-01' <= df['date']) & (df['date'] <= '2022-12-31')]['revenue'].sum()
```

```python
x = ['Q1', 'Q2', 'Q3', 'Q4']
y = [q1,q2,q3,q4]
```

```python
plt.bar(x,y, color = ['gray','b','c','g'], hatch = '/')
plt.title('Total revenue by quarter, 2022')
plt.show()
```

**Exercise 5 : Data Visualisation Using MatPlotLib**

```python
x = np.random.randint(-10,10,50)
x = pd.Series(x)
x = x.sort_values().values
y = x * x
```

```python
plt.plot(x,y)
plt.title('y = x ^ 2')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
products = ['A','B','C','D']
sales = [15, 30, 45, 20]

plt.bar(products, sales, color='red')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Week Sales')
plt.show()
```

```python
x = ['Apples', 'Bananas', 'Cherries', 'Dates']
y = [40, 30, 20, 10]

plt.figure(figsize=(10, 5))

plt.pie(y, labels = x, colors=['green', 'yellow', 'blue', 'red'])
plt.title('Favorite Fruits')
plt.legend(x)
plt.axis('equal')

plt.show()
```
