**Correcting any inconsistencies in the "Gender"**

```python
df["Gender"].unique()
```

```python
df['Gender'].replace(to_replace = [
       'M', 'Male', 'male', 'm', 'Male-ish', 'maile',
       'something kinda male?', 'Mal', 'Make', 'Male ', 'Man',
       'msle', 'Mail', 'Malr'], value = 'Male',
                     inplace = True)
```
