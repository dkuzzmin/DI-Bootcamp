**Let's make a column with salaries in USD**


```python
df1 = df[df['salary_currency'] == 'GBP']['salary'] * 1.27
df2 = df[df['salary_currency'] == 'EUR']['salary'] * 1.10
df3 = df[df['salary_currency'] == 'USD']['salary']
df_usd = pd.concat([df1, df2, df3])
df['salary_in_usd'] = df_usd
df[df['salary_currency'] == 'EUR']
```

**Perform a group-wise analysis of the experience_level column, calculating the mean and median salaries for each group using Pandas .groupby.**

```python
grouped = df.groupby('experience_level')['salary_in_usd'].agg(['mean', 'median'])
```

```python
grouped
```

**Utilize Pandas integration with Matplotlib to create a bar chart visualizing**

```python
categories = grouped.index
values = grouped['mean']
plt.xlabel('Experience Level')
plt.ylabel('Average Salary')
plt.bar(categories, values, color=['b', 'r', 'g', 'y'])
plt.title('Average Salary (by Experience Level)')

plt.show()
```
