# Exercise 2 : Scraping Robots.Txt From Wikipedia

```python
url = 'https://en.wikipedia.org/robots.txt'
response = urlopen(url)
k = response.read()
soup1 = BeautifulSoup(k, 'html.parser')
print(soup1.prettify())
```

# Exercise 6 : Scraping Movie Details

```python
url = 'https://editorial.rottentomatoes.com/guide/best-movies-of-2023/'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
n = soup.find_all(class_="article_movie_title", limit=10)


titles = [item.find('h2').find('a').get_text() for item in n]
titles
```

```python
years = [item.find('a').next.next.next.get_text() for item in n]
years
```

```python
n = soup.find_all(class_="info synopsis", limit=10)
briefs = [item.get_text().replace('Synopsis: ','').replace(' [More]','') for item in n]
briefs
```

```python
import pandas as pd
df = pd.DataFrame({'Title' : titles,
                                'Year' : years,
                                'Brief' : briefs },
                                columns=['Title','Year', 'Brief'])
df
```
