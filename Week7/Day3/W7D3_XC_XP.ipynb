# Exercise 1: Implementing And Comparing Normalization Techniques



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import seaborn as sns
from scipy import stats
```

```python
data = {
    'FeatureA' : np.random.randint(0, 1000, 10),
    'FeatureB' : np.random.randint(0, 10, 10)
}
df = pd.DataFrame(data)
df
```

```python
sns.scatterplot(df, x ='FeatureA', y = 'FeatureB')
```

```python
scaler = MinMaxScaler()

df_normalized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

print("Normalized Data:")
print(df_normalized)
```

```python
sns.scatterplot(df_normalized, x ='FeatureA', y = 'FeatureB')
```

```python
scaler = StandardScaler()
df_standardized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

print("Standardized Data:")
print(df_standardized)
```

```python
sns.scatterplot(df_standardized, x ='FeatureA', y = 'FeatureB')
```

Results: MinMax normalizing looks more appropriate for this simple dataset

# Exercise 3: Experimenting With Z-Score Normalization On Simulated Data

```python

data = {
    'FeatureX': np.random.normal(50, 10, 40),
    'FeatureY': np.random.normal(500, 100, 40)
}
df = pd.DataFrame(data)
df
```

```python
mean_X = np.mean(df['FeatureX'])
std_X = np.std(df['FeatureX'])

mean_Y = np.mean(df['FeatureY'])
std_Y = np.std(df['FeatureY'])

df_standardized1 = pd.DataFrame()

df_standardized1['FeatureX'] = (df['FeatureX'] - mean_X) / std_X
df_standardized1['FeatureY'] = (df['FeatureY'] - mean_Y) / std_Y

print("Standardized Data:")
print(df_standardized1)
```

```python
scaler = StandardScaler()
df_standardized2 = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

print("Standardized Data:")
print(df_standardized2)
```

```python

plt.figure(figsize=(16, 7))
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.set_title('Original')
ax1.hist(df['FeatureY'], bins=50, alpha=0.7, label='Original')
ax2.set_title('Z-score Normalization')
ax2.hist(df_standardized2['FeatureY'], bins=50, alpha=0.7, label='Standardized')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Normalized dataset's distribution remains the same as an original

# Exercise 5 : Handling Outliers In A Large Dataset

```python
df = pd.read_csv('AB_NYC_2019.csv')
df.head(10)
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]
```

```python
print(Q1)
print(Q3)
print(IQR)
```

```python
outliers
```

```python
trimmed_df = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
print(len(df))
print(len(trimmed_df))
```

We droped about 2400 records from dataset

```python
from scipy.stats.mstats import winsorize

winsorized_data = winsorize(df['price'], limits=[0.05, 0.05])

winsorized_df = pd.DataFrame(winsorized_data, columns=['price'])

print('Min:')
print(df['price'].min())
print(winsorized_df['price'].min())

print('Max:')
print(df['price'].max())
print(winsorized_df['price'].max())

```

We increased the minimum price value to 40 and decreased the maximum dataset value to 350

```python
# Log transformation
df['price_log'] = np.log(df['price'] + 1)
```

```python
print('Min:')
print(df['price'].min())
print(df['price_log'].min())

print('Max:')
print(df['price'].max())
print(df['price_log'].max())
```

```python
df[['price_log','price']].head(10)
```

Log transformation decreased the range of dataset from (0 : 10000) to (0 : 9.21)

The trimming method resulted in deleting too large amount of data. The winsorize method resulted in a good reduction in outliers. Log transformation seems ineffective for adjusting such a linear quantity as price

# Exercise 6 : Impact Of Outliers On Data Analysis

```python
df = pd.read_csv('WineQT.csv')
df.head()
```

```python
df[['fixed acidity', 'residual sugar', 'alcohol']].describe()
```

```python
plt.boxplot(df[['fixed acidity', 'residual sugar', 'alcohol']])
plt.show()
```

```python
Q1 = df['fixed acidity'].quantile(0.25)
Q3 = df['fixed acidity'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_acid = df[(df['fixed acidity'] < lower_bound) | (df['fixed acidity'] > upper_bound)]
trimmed_acid = df[(df['fixed acidity'] >= lower_bound) & (df['fixed acidity'] <= upper_bound)]
```

```python
print('Sum outliers:', len(outliers_acid))
print('Sum after:   ', len(trimmed_acid))
print('Original df: ', len(df))
mean1 = np.mean(df['fixed acidity'])
std1 = np.std(df['fixed acidity'])
mean2 = np.mean(trimmed_acid ['fixed acidity'])
std2 = np.std(trimmed_acid ['fixed acidity'])
print('Mean before:', mean1)
print('Mean after:', mean2)
print('St dev before:', std1)
print('St dev after:', std2)
```
