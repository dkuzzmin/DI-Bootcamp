```python
scaler = MinMaxScaler()

df[['fnlwgt_norm','capital-gain_norm','capital-loss_norm']] = pd.DataFrame(scaler.fit_transform(df[['fnlwgt','capital-gain','capital-loss']]), columns=['fnlwgt','capital-gain','capital-loss'])
```

```python
scaler = StandardScaler()
hours_per_week_2d = df['hours-per-week'].values.reshape(-1, 1)
df['hours-per-week_norm'] = pd.DataFrame(scaler.fit_transform(hours_per_week_2d))
```

```python
Q1 = df['capital-loss_norm'].quantile(0.25)
Q3 = df['capital-loss_norm'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers2 = df[(df['capital-loss_norm'] < lower_bound) | (df['capital-loss_norm'] > upper_bound)]
outliers2
```
